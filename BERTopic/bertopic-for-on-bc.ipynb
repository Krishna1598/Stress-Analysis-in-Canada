```python
#Defining path of 3 preprocessed files of before_covid, during_covid and after_covid for ON and BC.
BASE_PATH = "/kaggle/input/on-bc-preprossed-data/"

before_covid_path = BASE_PATH + "Final_before_covid_ON.csv"
during_covid_path = BASE_PATH + "Final_during_covid_ON.csv"
after_covid_path = BASE_PATH + "Final_after_covid_ON.csv"

before_covid_path = BASE_PATH + "Final_before_covid_BC.csv"
during_covid_path = BASE_PATH + "Final_during_covid_BC.csv"
after_covid_path = BASE_PATH + "Final_after_covid_BC.csv"
```

```python
#Reading 3 files 
before_covid = pd.read_csv(before_covid_path)
during_covid = pd.read_csv(during_covid_path)
after_covid = pd.read_csv(after_covid_path)
```

```python
list_of_words_before_covid = before_covid['TWEET'].to_list()
list_of_words_during_covid = during_covid['TWEET'].to_list()
list_of_words_after_covid = after_covid['TWEET'].to_list()
```

```python
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
```

```python
#before COVID-19
topics_before, probs_before = topic_model.fit_transform(list_of_words_before_covid)
#during COVID-19
topics_during, probs_during = topic_model.fit_transform(list_of_words_during_covid)
#after COVID-19
topics_after, probs_after = topic_model.fit_transform(list_of_words_after_covid)
```
